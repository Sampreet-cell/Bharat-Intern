```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
from PIL import Image
```

# Load the MNIST (Modified National Institute of Standards and Technology) dataset

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

# Data Preprocessing by:-

```python
x_train = x_train.reshape(-1, 28, 28, 1) / 255.0
x_test = x_test.reshape(-1, 28, 28, 1) / 255.0
```

### 1. Conversion to one-hot encoding

```python
y_train = tf.keras.utils.to_categorical(y_train, num_classes=10)
y_test = tf.keras.utils.to_categorical(y_test, num_classes=10)
```

### 2. Data augmentation

```python
datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1
)
datagen.fit(x_train)
```

# Building and compiling the CNN (Convolutional Neural Network) model

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

## Training the model with the augmented data

```python
model.fit(datagen.flow(x_train, y_train, batch_size=128), epochs=200, validation_data=(x_test, y_test))
```

# Save the model and load it

```python
model.save("mnist_cnn_model.h5")
```

```python
model = load_model("mnist_cnn_model.h5")
```

# User's image input for prediction

```python
image_path = "sample1.jpg"  # Replace with the path to the user's image
image = Image.open(image_path).convert('L')  # Open the image and convert to grayscale
image = image.resize((28, 28))  # Resize the image to match the MNIST dataset
input_data = np.array(image) / 255.0  # Preprocess the image data
input_data = input_data.reshape(1, 28, 28, 1) 
```

# Final Prediction

```python
predictions = model.predict(input_data)
predicted_label = np.argmax(predictions[0])  # Get the predicted label
print(f"Predicted digit: {predicted_label}")
```
